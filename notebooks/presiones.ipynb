```python
import matplotlib.pyplot as plt
import numpy as np

from diving_calc.physics.depth_converter import DepthConverter
from diving_calc.physics.pressure_converter import Density, Gravity, AltitudePressure, PressureConverter
```

```python
altitud = np.linspace(0, 3000, 3000)
profundidad = [10, 20, 30, 40]
def profundidad_a_atm(altitud, depth):
    conv = DepthConverter.for_fresh_water(altitud)
    return conv.to_atm(depth)
funcion_vectorizada = np.vectorize(profundidad_a_atm)
```

```python
plt.figure(figsize=(8, 5))
for c in profundidad:
    y = funcion_vectorizada(altitud, c)  # Aplicar la función a cada x
    plt.plot(altitud, y, label=f'-{c}m')
plt.xlabel('m')
plt.ylabel('ATM')
plt.title('Presion segun la altitud')
plt.legend()
plt.grid()
plt.show()
```
